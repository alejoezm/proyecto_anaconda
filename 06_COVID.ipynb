```python
for fecha in ['fecha reporte web','Fecha de notificación','Fecha de inicio de síntomas','Fecha de muerte', 'Fecha de diagnóstico', 'Fecha de recuperación',]:
    covid[fecha]=pd.to_datetime(covid[fecha], format='%d/%m/%Y %H:%M:%S')
```

```python
covid.info()
```

```python
#acumulado de casos por dias
cf=covid.groupby(['Fecha de diagnóstico'])[['Edad']].describe()
cf.columns=cf.columns.droplevel(0)
cf
```

```python
fig, ax=plt.subplots(figsize=(16,6))
ax.plot(cf.index, cf['count'], color='green', alpha=0.4)
ax.set(title='total_pacientes_diagnosticados por x dia', xlabel='Fechas', ylabel='# pacientes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
cf['acumulados']=cf['count'].cumsum()
fig, ax=plt.subplots(figsize=(16,6))
ax.plot(cf.index, cf['acumulados'], 'r:')
ax.set(title='Pacientes acumulados por fecha', xlabel='Fechas', ylabel='# pacientes')
#plt.xticks(rotation=45, ha='right')
ax.grid()
plt.show()
```
